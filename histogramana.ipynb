```python
def datapreparet(datain):
    dataa=np.array(datain[datain.columns[1:13]])
    labela=np.array(datain[datain.columns[-1]])
    indn1=np.where(dataa[:,2]>0.5)[0]
    dataa1=dataa[indn1,:]
    labela1=labela[indn1]
    indn2=np.where(dataa1[:,6]>0.5)[0]
    dataa2=dataa1[indn2,:]
    labela2=labela1[indn2]
    indn3=np.where(dataa2[:,10]>0.5)[0]
    dataa3=dataa2[indn3,:]
    labela3=labela2[indn3]
    indn4=np.where(dataa3[:,1]<3500)[0]
    dataa4=dataa3[indn4,:]
    labela4=labela3[indn4]
    indn5=np.where(dataa4[:,5]<3500)[0]
    dataa5=dataa4[indn5,:]
    labela5=labela4[indn5]
    indn6=np.where(dataa5[:,9]<3500)[0]
    dataa6=dataa5[indn6,:]
    labela6=labela5[indn6]
    indn7=np.where(labela6!=12833)[0]
    dataa7=dataa6[indn7,:]
    labela7=labela6[indn7]
    indn8=np.where(labela7!=9507)[0]
    dataa8=dataa7[indn8,:]
    labela8=labela7[indn8]
    labelin=labela8
    datain=dataa8
    
    
 
    all_feature=datain
    all_label=labelin
    return all_feature, all_label
```

```python
def metrixout(datain):
    matrixout=np.zeros((len(datain),len(datain)))
    for i in range(len(datain)):
        data1=datain[i]
        for j in range(len(datain)):
            data2=datain[j]
            metric_val = np.corrcoef(data1.flatten(), data2.flatten())[0,1]
            matrixout[i,j]=metric_val
    return matrixout
```

```python
def matrixtodata(matrixin):
    output=[]
    for i in range(len(matrixin)):
        for j in range(len(matrixin)):
            if j>i:
                output.append(matrixin[i,j])
    outputa=np.array(output)
    return outputa
```

```python
def clustering(datam):
    kmeansm=KMeans(n_clusters=10,random_state=42)
    kmeansm=kmeansm.fit(datam)
    labelsm=kmeansm.labels_
    centersm=kmeansm.cluster_centers_
    return labelsm,centersm
```

```python
def featuresep(featurein):
    scaler=MinMaxScaler()
    meanout=scaler.fit_transform(np.c_[featurein[:,0],featurein[:,4],featurein[:,8]])
    rmsout=scaler.fit_transform(np.c_[featurein[:,1],featurein[:,5],featurein[:,9]])
    stdout=scaler.fit_transform(np.c_[featurein[:,2],featurein[:,6],featurein[:,10]])
    fmeanout=scaler.fit_transform(np.c_[featurein[:,3],featurein[:,7],featurein[:,11]])
    return meanout, rmsout, stdout, fmeanout
```

```python
def plotall(pathin,datain,namein):
    fig=plt.figure(figsize=(30,16))
    num_rows = len(datain)//7+(len(datain)%7>0)
    for i,file_name in enumerate(datain):
        full_path=pathin+'/'+file_name
        plt.subplot(num_rows,7,i+1)
        
        img=plt.imread(full_path)
        plt.imshow(img,cmap=plt.cm.gray)
        #plt.xticks([])
        #plt.yticks([])
        plt.axis('off')
    plt.subplots_adjust(wspace=0.1)
    
    plt.tight_layout()
    plt.show()
    fig.savefig(namein,bbox_inches='tight', pad_inches=0,dpi=300)
```
