```python
import sklearn
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
```

```python
def featurec(path,files):
    t=np.zeros((len(files),10))
    for i in range(len(files)):
        x,sr1=librosa.load(path+"\\"+files[i],sr=None)
        duration=librosa.get_duration(y=x, sr=sr1)
        emphasized_signal = np.append(x[0], x[1:] -0.97 * x[:-1])
        mfccss = librosa.feature.mfcc(y=emphasized_signal, sr=sr1)
        mfccs = normalize(mfccss , axis=1)
        mfccs=mfccs.T
        mfccsm=np.mean(mfccs)
        f0, voiced_flag, voiced_probs = librosa.pyin(x,sr=sr1,frame_length=512, fill_na=np.nan, n_thresholds=50, fmin=120, fmax=600)
        rf0= np.nanmax(f0)-np.nanmin(f0)
        if np.isnan(rf0):
            print(path+"\\"+files[i])
        maxf0 = np.nanmax(f0)
        minf0 = np.nanmin(f0)
        mmf0=np.nanmean(f0)
        rms = librosa.feature.rms(y=x,frame_length=512, hop_length=128)
        mrms=np.nanmean(rms)
        zerocrossr = librosa.feature.zero_crossing_rate(y=x,frame_length=512, hop_length=128)
        mzerocrossr=np.nanmean(zerocrossr)
        starttime=re.search("(\d+\.\d+)",files[i]).group(0)
        label=0
        feature=np.c_[float(starttime),duration,mfccsm,rf0,maxf0,minf0,mmf0,mrms,mzerocrossr,label]
        t[i,:]=feature
    return t
```

```python
def featureci(path,files):
    t=np.zeros((len(files),10))
    for i in range(len(files)):
        x,sr1=librosa.load(path+"\\"+files[i],sr=None)
        duration=librosa.get_duration(y=x, sr=sr1)
        emphasized_signal = np.append(x[0], x[1:] -0.97 * x[:-1])
        mfccss = librosa.feature.mfcc(y=emphasized_signal, sr=sr1)
        mfccs = normalize(mfccss , axis=1)
        mfccs=mfccs.T
        mfccsm=np.mean(mfccs)
        f0, voiced_flag, voiced_probs = librosa.pyin(x,sr=sr1,frame_length=512, fill_na=np.nan, n_thresholds=50, fmin=120, fmax=1000)
        rf0= np.nanmax(f0)-np.nanmin(f0)
        if np.isnan(rf0):
            print(path+"\\"+files[i])
        maxf0 = np.nanmax(f0)
        minf0 = np.nanmin(f0)
        mmf0=np.nanmean(f0)
        rms = librosa.feature.rms(y=x,frame_length=512, hop_length=128)
        mrms=np.nanmean(rms)
        zerocrossr = librosa.feature.zero_crossing_rate(y=x,frame_length=512, hop_length=128)
        mzerocrossr=np.nanmean(zerocrossr)
        starttime=re.search("(\d+\.\d+)",files[i]).group(0)
        label=1
        feature=np.c_[float(starttime),duration,mfccsm,rf0,maxf0,minf0,mmf0,mrms,mzerocrossr,label]
        t[i,:]=feature
    return t
```

```python
def diffc(pathturn,filesturn,pathturnc,filesturnc):
    t=featurec(pathturn,filesturn)
    tc=featureci(pathturnc,filesturnc)
    block=np.r_[t,tc]
    blocks = block[block[:,0].argsort()]
    groups = [(k, sum(1 for _ in g)) for k, g in groupby(blocks[:,-1])]
    groupn=np.zeros((len(groups),1))
    for i in range(len(groups)):
        groupn[i]=groups[i][1]
    
    ind=np.zeros((len(groupn),2))
    for i in range(len(groupn)):
        if i!=len(groupn):
            ind[i,:]=np.c_[sum(groupn[0:i]),sum(groupn[0:i+1])]
    blockn=np.zeros((len(groupn),10))
    for i in range(len(groupn)):
        blockn[i,:]=np.mean(blocks[int(ind[i,0]):int(ind[i,1]),:],axis=0)
    diff=np.zeros((len(groupn)-1,10))
    mother=np.zeros((len(groupn)-1,10))
    infant=np.zeros((len(groupn)-1,10))
    for i in range(len(groupn)-1):
        diff[i,:]=np.absolute(blockn[i+1,:]-blockn[i,:])
        if (blockn[i+1,-1]==0)&(blockn[i,-1]==1):
            mother[i,:]=blockn[i+1,:]
            infant[i,:]=blockn[i,:]
        if (blockn[i+1,-1]==1)&(blockn[i,-1]==0):
            mother[i,:]=blockn[i,:]
            infant[i,:]=blockn[i+1,:]
    xlabel=(np.arange(len(diff))+1)
    print(len(diff))
    return -diff,xlabel,mother,infant
```

```python
def remove16(datain):
    labelstart=np.where(datain[:,1]==1)[0]
    labelend=np.zeros((len(labelstart)))
    labelend[:len(labelstart)-1]=labelstart[1:len(labelstart)]-1
    labelend[-1]=len(datain[:,1])-1
    endindex=datain[labelend.astype(int),1]
    startr=labelstart[np.where(endindex.astype(int)==16)[0]].astype(int)
    #print(startr)
    endr=labelend[np.where(endindex.astype(int)==16)[0]].astype(int)
    #print(endr)
    dataformer=datain[:startr[0]]
    datalatter=datain[endr[0]+1:]
    dataout=np.r_[dataformer,datalatter]
    #print(len(datain),len(dataout),dataout[:,1])
    return dataout
```

```python
# Define different types of functions to fit
def linear_func(x, a, b):
    return a * x + b

def quadratic_func(x, a, b, c):
    return a * x**2 + b * x + c

def cubic_func(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

def four_func(x, a, b, c, d,e):
    return a * x**4 + b * x**3 + c * x**2 + d*x+e

def five_func(x, a, b,c,d,e,f):
    return a * x**5 + b * x**4 + c * x**3 + d*x**2+e*x+f

def six_func(x, a, b,c,d,e,f,g):
    return a * x**6 + b * x**5 + c * x**4 + d * x**3 + e * x**2 + f * x + g
def seven_func(x, a, b,c,d,e,f,g,h,i):
    return a * x**8 + b * x**7 + c * x**6 + d * x**5 + e * x**4 + f * x**3 + g*x**2+h*x+i
#def sigmoid_func(x, L, k, x0):
    #return a * x**3 + b * x**2 + c * x + d

#def gaussian_func(x, a, x0, sigma):
    #return return a * x**3 + b * x**2 + c * x + d
```

```python
# Curve fitting for each function
functions = [linear_func, quadratic_func, cubic_func, four_func,
             five_func]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
def calculate_bic(y_true, y_pred, n, k): #Bayesian Information Criterion
    mse = mean_squared_error(y_true, y_pred)
    bic = n * np.log(mse) + k * np.log(n)
    return bic
```

```python
def outputfit(functionsin,xin,yin):
    r2_values=[]
    fit_results=[]
    iii=2
    for func in functionsin:
        popt, _ = curve_fit(func, xin, yin)
        fit_results.append(popt)
        y_fit = func(xin, *popt)
        r2_values.append(calculate_bic(yin, y_fit, len(yin), iii))
        iii=iii+1
    return r2_values,fit_results
```
