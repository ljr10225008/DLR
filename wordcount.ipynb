```python
import os
import numpy as np
import pandas as pd
```

```python
idpath = r"F:\DLR\wordcounttimealign\6mALICE"
idlist= os.listdir(idpath)
```

```python
idlist.sort()
print(idlist)
```

```python
wordc=np.zeros((len(idlist),12))
```

```python
def kchi(file_name):
    num=0
    numa=0
    i=0 
    with open(file_name, 'r', encoding='utf-8') as f: 
        lines = f.readlines()
        line_n=len(lines)
        startp=np.zeros((line_n,1))
        duration=np.zeros((line_n,1))
        who=[np.nan]*line_n
    with open(file_name, 'r', encoding='utf-8') as f:
        for line in f:
            line=line.split(' ')
            startp[i]=line[3] #start of an utterance
            duration[i]=line[4] #duration of an utterance
            who[i]=line[7] #who is the speaker of an utterance
            if who[i]=='KCHI':
                num=num+1
            if who[i]=='FEM'or who[i]=='MAL':
                numa=numa+1
            i=i+1
    return num,numa
```

```python
def dur(file_name):
    durout=0
    i=0 
    with open(file_name, 'r', encoding='utf-8') as f: 
        lines = f.readlines()
        line_n=len(lines)
        startp=np.zeros((line_n,1))
        duration=np.zeros((line_n,1))
        who=[np.nan]*line_n
    with open(file_name, 'r', encoding='utf-8') as f:
        for line in f:
            line=line.split(' ')
            duration[i]=line[4] #duration of an utterance
            who[i]=line[7] #who is the speaker of an utterance
            if who[i]=='FEM'or who[i]=='MAL':
                durout=durout+duration[i]
            i=i+1
    return durout
```

```python
def dur2(file_name):
    durouta=0
    durouti=0
    i=0 
    a=0
    c=0
    with open(file_name, 'r', encoding='utf-8') as f: 
        lines = f.readlines()
        line_n=len(lines)
        startp=np.zeros((line_n,1))
        duration=np.zeros((line_n,1))
        who=[np.nan]*line_n
    with open(file_name, 'r', encoding='utf-8') as f:
        for line in f:
            line=line.split(' ')
            duration[i]=line[4] #duration of an utterance
            who[i]=line[7] #who is the speaker of an utterance
            if who[i]=='FEM'or who[i]=='MAL':
                durouta=durouta+duration[i]
                a=a+1
            if who[i]=='KCHI':
                durouti=durouti+duration[i]
                c=c+1
            i=i+1
    return durouta,durouti,a,c
```

```python
for i in range(len(idlist)):
    aaa=idlist[i]
    wordc[i,0]=int(aaa)
    p=0
    s=0
    w=0
    u_i=0
    u_a=0
    path=r"F:\DLR\wordcounttimealign\6mALICE"+"\\"+aaa
    files=os.listdir(path)
    hour=len(files)/2
    #print(hour)
    for j in range(len(files)):
        filename=path+"\\"+files[j]+"\\"+'ALICE_output.txt'
        sfilename=path+"\\"+files[j]+"\\"+'diarization_output.rttm'
        df=pd.read_table(filename)
        p=p+int(df[' phonemes '])
        s=s+int(df[' syllables '])
        w=w+int(df[' words'])
        ui,ua=kchi(sfilename)
        u_a=u_a+ua
        u_i=u_i+ui
    wordc[i,1]=p
    wordc[i,2]=s
    wordc[i,3]=w
    wordc[i,4]=u_a
    wordc[i,5]=u_i
    wordc[i,6]=hour
    wordc[i,7]=p/hour
    wordc[i,8]=s/hour
    wordc[i,9]=w/hour
    wordc[i,10]=u_a/hour
    wordc[i,11]=u_i/hour
```

```python
proout=np.zeros((len(idlist),2))
for i in range(len(idlist)):
    aaa=idlist[i]
    proout[i,0]=int(aaa)
    pc=0
    path=r"F:\DLR\wordcounttimealign\6mALICE"+"\\"+aaa
    files=os.listdir(path)
    hour=len(files)/2
    #print(hour)
    for j in range(len(files)):
        sfilename=path+"\\"+files[j]+"\\"+'diarization_output.rttm'
        durc=dur(sfilename)
        pc=pc+durc
    proout[i,1]=pc/(hour*3600)
```

```python
proout2=np.zeros((len(idlist),3))
for i in range(len(idlist)):
    aaa=idlist[i]
    proout2[i,0]=int(aaa)
    pa2=0
    aa2=0
    pc2=0
    cc2=0
    path=r"F:\DLR\wordcounttimealign\6mALICE"+"\\"+aaa
    files=os.listdir(path)
    #print(hour)
    for j in range(len(files)):
        sfilename=path+"\\"+files[j]+"\\"+'diarization_output.rttm'
        dura2,durc2,a2,c2=dur2(sfilename)
        pa2=pa2+dura2
        aa2=aa2+a2
        pc2=pc2+durc2
        cc2=cc2+c2
    proout2[i,1]=pa2/aa2
    proout2[i,2]=pc2/cc2
```

```python
allsp=pd.DataFrame(proout2)
allsp.columns=['ID','dura_m','durc_m']
pathsave=r"F:\DLR\wordcountoutputs"
allsp.to_csv(pathsave+"\\"+'6moutdur.csv')
```

```python
allsp=pd.DataFrame(proout)
allsp.columns=['ID','pro']
pathsave=r"F:\DLR\wordcountoutputs"
allsp.to_csv(pathsave+"\\"+'6moutpro.csv')
```
