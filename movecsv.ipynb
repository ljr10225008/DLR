```python
import os
import shutil
```

```python
def move_rename(source_folder,target_folder,filename='VTC.csv'):
    if not os.path.exists(source_folder):
        raise FileNotFoundError(f"The source folder '{source_folder}' does not exist.")
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    source_file_path=os.path.join(source_folder,filename)
    if not os.path.exists(source_file_path):
        raise FileNotFoundError(f"The file '{filename}' does not exist in the folder '{source_folder}'.")
    original_folder_name = os.path.basename(source_folder)
    new_file_name = f"{original_folder_name}{filename}"
    target_file_path=os.path.join(target_folder,new_file_name)
    
    shutil.move(source_file_path,target_file_path)
    #print(f"File moved and renamed to: {target_file_path}")
```

```python
path_parent=r'F:\DLR\matchedspeakersound\new_0730\text\18m'
folders_parent=os.listdir(path_parent)
folders_parent.sort()
```

```python
path_target=r'F:\DLR\matchedspeakersound\new_0730\speaker\18m'
```

```python
for i in range(len(folders_parent)):
    source_folders=path_parent+'\\'+folders_parent[i]
    folders_source=os.listdir(source_folders)
    folders_source.sort()
    target=path_target+'\\'+folders_parent[i]
    for j in range(len(folders_source)):
        sourcein=source_folders+"\\"+folders_source[j]
        move_rename(sourcein,target)
```
